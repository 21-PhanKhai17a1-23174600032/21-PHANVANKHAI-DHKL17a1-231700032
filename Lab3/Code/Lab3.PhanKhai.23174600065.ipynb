LAB 3.2. XỬ LÝ DỮ LIỆU NULL (thời gian thực hiện: 15-20 phút.)
Mục tiêu: Học cách xác định và xử lý dữ liệu Null.
Nhiệm Vụ:
1. Kiểm tra xem trong stocks1 có dữ liệu Null nào không.
2. Thay thế dữ liệu Null ở cột high bằng giá trị trung bình của cột high.
3. Thay thế dữ liệu Null ở cột low bằng giá trị trung bình của cột low.
4. Hiển thị thông tin tổng quan để xác nhận không còn dữ liệu Null.

```python

#Kiểm tra dữ liệu Null trong stocks1
Ktra_Null= stocks1.isnull()
print(Ktra_Null.sum()) 

#Tính gtri trun bình và thay thế gtri Null
gtri_tb=stocks1['high'].mean()
stocks1['high'] = stocks1['high'].fillna(gtri_tb)
print(stocks1['high'])


#Tính gtri trun bình và thay thế gtri Null
gtri_tbinh=stocks1['low'].mean()
stocks1['low'] = stocks1['low'].fillna(gtri_tbinh)
print(stocks1['low'])

#Hiển thị tổng quan sau khi thay thế gtri
stocks1.info()
```

LAB 3.3. GỘP DỮ LIỆU VÀ PHÂN TÍCH CƠ BẢN (thời gian thực hiện: 15-20 phút.)
Mục tiêu: Nắm vững cách gộp dữ liệu và thực hiện phân tích cơ bản.
Nhiệm Vụ:
1. Đọc file stocks2.csv vào DataFrame stocks2.
2. Gộp stocks1 và stocks2 thành DataFrame mới tên là stocks.
3. Tính giá trung bình (open, high, low, close) cho mỗi ngày.
4. Hiển thị 5 dòng đầu tiên của kết quả.


```python
file='D:/Lab3/Data/stocks2.csv'
stocks2=pd.read_csv(file)

#Gộp stocks1 và stocks2
stocks=pd.concat([stocks1,stocks2],ignore_index=True)
print(stocks)

#Tính gtri trung bình cho mỗi ngày
stocks['average'] = stocks[['open', 'high', 'low', 'close']].mean(axis=1)
print(stocks)
stocks[:5]
```

LAB 3.4. KẾT HỢP DỮ LIỆU TỪ NHIỀU NGUỒN (thời gian thực hiện: 15-20 phút.)
Mục Tiêu: Nắm được cách kết hợp dữ liệu từ các file khác nhau và thực hiện phân tích cơ
bản.
Nhiệm Vụ:
1. Đọc file companies.csv vào DataFrame companies.
2. Hiển thị 5 dòng đầu tiên của companies.
3. Kết hợp stocks (đã tạo từ bài 3) và companies dựa trên cột chung là symbol.
4. Tính giá đóng cửa (close) trung bình cho mỗi công ty.
5. Hiển thị kết quả cho 5 công ty đầu tiên.


```python
file_company='D:/Lab3/Data/companies.csv'
companies=pd.read_csv(file_company)
print(companies[:5])

#kết hợp theo cột chung là symbol
data=pd.merge(stocks, companies,left_on='symbol', right_on='name')
print(data)

#Tính trung bình giá close của mỗi công ty
data.groupby('name')['close'].mean()
print(data)

#Hiển thị kqua cho 5 cty đầu tiên 
print(data[:5])
```

LAB 3.5: SỬ DỤNG MULTIINDEX VÀ GROUPBY(thời gian thực hiện: 15-20 phút.)
Mục Tiêu: Hiểu cách sử dụng MultiIndex và GroupBy để phân tích dữ liệu một cách hiệu
quả.
Nhiệm Vụ:
1. Tạo MultiIndex cho DataFrame stocks bằng cách sử dụng cột date và symbol làm
chỉ mục.
2. Sử dụng GroupBy để tính giá trung bình (open, high, low, close) và volume trung
bình cho mỗi ngày, cho mỗi mã chứng khoán.
3. Sắp xếp dữ liệu theo ngày và mã chứng khoán.
4. Hiển thị kết quả cho 5 ngày đầu tiên.

```python
#Tạo multiindex cho dataframe 
stocks_index=stocks.set_index(['symbol', 'date'])
print(stocks_index)

#Sử dụng groupby để tính gtri trung bình
group = stocks_index.groupby(level='symbol')
result = group[['open', 'high', 'low', 'close']].mean()
print(result)

#Sắp xếp dữ liệu theo ngày và mã chứng khoán
sap_xep=stocks_index.sort_index(level=['symbol','date'])
print(sap_xep)

#Hiển thị 5 ngày đầu tiên
print(sap_xep[:5])
```
